```python
import vmodel
import os
import numpy as np
import h5py
import datetime
import scipy.spatial
import math
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import seaborn as sns; sns.set_theme()
import pandas as pd
```

```python
pol = np.load("/home/lars/vmodel_output/test_mill_longsim_pol_astr_dphi.npy")

steps = 10
reps = 20

paraChange1_val = np.linspace(0,3,steps)
paraChange2_val = np.linspace(0,1,steps)

pol = np.mean(pol, axis = 3)
var_pol = np.var(pol, axis = 2)
pol = np.mean(pol, axis = 2)

critline_idx = []
critline = []
for i in range(steps):
    critline_idx.append(np.argmax(var_pol[:,i]))
    critline.append(paraChange1_val[np.argmax(var_pol[:,i])])

df = pd.DataFrame(np.flip(pol, axis = 0))

df.columns = np.round(paraChange2_val,2)
df.index = np.round(paraChange1_val,2)[::-1]


#plt.title(figname, fontsize = 12, pad = 20) # title with fontsize 20
ax = sns.heatmap(df, cbar_kws={'label': "polarization $Φ$"},
                 cmap="Reds")
plt.plot(abs(np.array(critline)-3)*3, color="purple", linewidth=3, linestyle = "--")
plt.tight_layout()
plt.gcf().set_size_inches(10,7.85)
plt.xlabel("angular noise $D$", fontsize = 15) # x-axis label with fontsize 15
plt.ylabel("alignment strength " + u'$μ_{alg}$', fontsize = 15) # y-axis label with fontsize 15
plt.title("Occluded, field of vision = 330 °")
plt.savefig("HM_pol_occ_dphi_critline.pdf",bbox_inches="tight")
```
