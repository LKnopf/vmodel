```python
import vmodel
import os
import numpy as np
```

```python
paraChange1_name = "--alignment-strength"
paraChange2_name = "--dphi"
steps = 2
time = 10
reps = 4
saveLoc = "/home/lars/vmodel_output/test2"

paraChange1_val = np.linspace(0,1.5,steps)
paraChange2_val = np.linspace(0,1,steps)



if (not saveLoc.endswith('/')):
    saveLoc += "/"
    
try:
    os.makedirs(saveLoc)
except FileExistsError:
    # directory already exists
    pass
```

```python
for i in paraChange1_val:
    for j in paraChange2_val:
        
        cmd = "vmodel --num-timesteps 15 --verbose --progress --filter-occluded --num-agents 100 --delta-time 0.02 --num-preds 1 --pred-time 2 --vision-pred 300 --flee-strength 0 --num-runs "+str(reps)+" --outfolder "+str(saveLoc)+" "+str(paraChange1_name)+" "+str(i)+" "+str(paraChange2_name)+" "+str(j)
        print(cmd)
        os.system(cmd)
        
```
