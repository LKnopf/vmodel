```python
import pickle
import pandas as pd
import numpy as np

with open('/home/lars/Downloads/BB36_mob_m1_frame00000--0001_dtc_converted.pkl', 'rb') as file:
    data = pickle.load(file)

dataFull = np.zeros((1135, 271))
dataFull[:] = np.nan
dataFull[:,0] = np.array(range(1135))+1

for i in range(135):
    try:
        entries = data[str(i+1)]["FRAME_IDX"]
        for idx in range(len(entries)):
            #print(idx)
            dataFull[entries[idx]-1,2*i+1] = data[str(i+1)]["X"][idx]
            dataFull[entries[idx]-1,2*i+2] = data[str(i+1)]["Y"][idx]
    except:
            0
    
        

COLNAM = []
COLNAM.append("FRAME_IDX")

for i in range(135):
    COLNAM.append("X_"+str(i+1))
    COLNAM.append("Y_"+str(i+1))
    
data_df = pd.DataFrame(dataFull, columns = COLNAM)


data_df.to_csv("data_df.csv")
np.savetxt("data_array.csv", dataFull, delimiter=",")
```
